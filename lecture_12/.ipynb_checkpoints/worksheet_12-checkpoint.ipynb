Underfitting means using a model that is too simple and therefore makes many mistakes. This would appear as a model with a high amount of mistakes while training. For this question we will re-use the "mnist_784" dataset.

a) Begin by creating a training and testing datasest from our dataset, with a 80-20 ratio, and random_state=1. You can use the `train_test_split` function from sklearn. By holding out a portion of the dataset we can evaluate how our model generalizes to unseen data (i.e. data it did not learn from).

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_openml

X, y = fetch_openml(name="mnist_784", version=1, return_X_y=True, as_frame=False)
training, testing, ytrain, ytest = train_test_split(X[:400], y[:400], test_size=0.2, random_state=1)
```

b) For K ranging from 1 to 20:

1. train a KNN on the training data
2. record the training and testing accuracy

Plot a graph of the training and testing set accuracy as a function of the number of neighbors K (on the same plot). Which value of K is optimal? Briefly explain.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

train_score = [0 for num in range(21)]
test_score = [0 for num in range(21)]

for i in range(1, 21):
    neighbor = KNeighborsClassifier(n_neighbors=i)
    neighbor.fit(training, ytrain)
    
    training_p = neighbor.predict(training)
    testing_p = neighbor.predict(testing)
    
    train_a = accuracy_score(ytrain, training_p)
    test_a = accuracy_score(ytest, testing_p)
    
    train_score[i] = train_a
    test_score[i] = test_a

plt.plot(range(21),train_score)
plt.plot(range(21),test_score)
plt.xlabel('K value - Number of Neighbors',size=20)
plt.ylabel('Accuracy Score',size=20)
plt.show()
    
```

The value of K that is optimal is 4. This is the value with the highest testing accuracy, meaning that the model works well for new data that is has not seen before. K=1 is excluded from consideration, despite having a high accuracy, because having 1 neighbor means that each point is its own neighbor.
